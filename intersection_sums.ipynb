# True-predicted intersections
Here I take a look at how the average counts of aggregated true and pred sets look like, as well as the intersections between the two (this is what the metric scores are based on)

I conclude that the distributions look as expected, without a significant difference between intersection and true/pred sums (which I feared)

```python
import dill
import matplotlib.pyplot as plt
import pandas as pd

from jupyterthemes import jtplot
jtplot.style()
```

```python
with open("intersection_sums_bert_extTrue.pickle", "rb") as f:
    sums = dill.load(f)
```

```python
intersection_sums, true_sums, predicted_sums = sums[0], sums[1], sums[2]
```

```python
print(len(intersection_sums))
print(len(true_sums))
print(len(predicted_sums))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15,15))

df1 = pd.DataFrame(intersection_sums, columns=["Intersection sums"])
bplot1 = df1.boxplot(ax=axes[0,0], showfliers=True)
bplot1.set_ylabel('Sums')

df2 = pd.DataFrame(predicted_sums, columns=["Predicted sums"])
bplot2 = df2.boxplot(ax=axes[0,1], showfliers=True)
bplot2.set_ylabel('Sums')

df3 = pd.DataFrame(true_sums, columns=["True sums"])
bplot3 = df3.boxplot(ax=axes[0,2], showfliers=True)
bplot3.set_ylabel('Sums')

df4 = pd.DataFrame(intersection_sums, columns=["Intersection sums"])
bplot1 = df4.boxplot(ax=axes[1,0], showfliers=False)
bplot1.set_ylabel('Sums')

df5 = pd.DataFrame(predicted_sums, columns=["Predicted sums"])
bplot1 = df5.boxplot(ax=axes[1,1], showfliers=False)
bplot1.set_ylabel('Sums')

df6 = pd.DataFrame(true_sums, columns=["True sums"])
bplot1 = df6.boxplot(ax=axes[1,2], showfliers=False)
bplot1.set_ylabel('Sums')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15,15))

df1 = pd.DataFrame(intersection_sums, columns=["Intersection sums"])
plot1 = df1.plot(ax=axes[0])

df2 = pd.DataFrame(predicted_sums, columns=["Predicted sums"])
plot2 = df2.plot(ax=axes[1])
plt.show()
```
